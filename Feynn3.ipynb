# 1. Data **Collection** and Processing
First, we'll collect and process data to use for recommendations and sales forecasting.

**bold text**

```python
import pandas as pd

# Load customer purchase history dataset
purchase_data = pd.read_csv('customer_purchase_history.csv')

# Basic preprocessing
purchase_data.dropna(inplace=True)
purchase_data['purchase_date'] = pd.to_datetime(purchase_data['purchase_date'])
```

```python
purchase_data.head()
```

Exploratory Data Analysis (EDA)
# Perform EDA to understand the data better and identify patterns.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the number of purchases over time
plt.figure(figsize=(10, 6))
purchase_data['purchase_date'].value_counts().sort_index().plot()
plt.title('Number of Purchases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Purchases')
plt.show()

# Distribution of purchases per customer
plt.figure(figsize=(10, 6))
sns.histplot(purchase_data['customer_id'].value_counts(), bins=50, kde=True)
plt.title('Distribution of Purchases Per Customer')
plt.xlabel('Number of Purchases')
plt.ylabel('Frequency')
plt.show()
```

# 3. Model Development
a. Recommendation System using Collaborative Filtering

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import cross_validate

# Load data into Surprise dataset
reader = Reader(rating_scale=(1, 5))

data = Dataset.load_from_df(purchase_data[['customer_id', 'product_id', 'rating']], reader)

# Use SVD for collaborative filtering
algo = SVD()

# Perform cross-validation
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)

# Train on the full dataset
trainset = data.build_full_trainset()
algo.fit(trainset)

# Predicting ratings for a specific user
user_id = 'A1'
item_id = 'B001'
prediction = algo.predict(user_id, item_id)
print(f'Predicted rating for user {user_id} and item {item_id}: {prediction.est}')
```

# b. Sales Forecasting using Time-Series Analysis


```python
from prophet import Prophet
import pandas as pd

# Assuming purchase_data DataFrame is already loaded
# Prepare data for Prophet
sales_data = purchase_data.groupby('purchase_date').size().reset_index(name='sales')
sales_data.rename(columns={'purchase_date': 'ds', 'sales': 'y'}, inplace=True)

# Initialize and train the model
model = Prophet()
model.fit(sales_data)

# Make future predictions
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# Plot the forecast
fig = model.plot(forecast)
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

# 4. Integration into Web Application
We will use Flask to create a simple web application to interact with the AI models.

```python
from flask import Flask, request, jsonify
import joblib

app = Flask(__name__)

# Load pre-trained models
# Make sure the file paths are correct and the files exist
recommendation_model = joblib.load('/content/recommendation_model.pkl') # Update with actual path
forecast_model = joblib.load('/content/forecast_model.pkl') # Update with actual path

@app.route('/recommend', methods=['GET'])
def recommend():
    user_id = request.args.get('user_id')
    item_id = request.args.get('item_id')
    prediction = recommendation_model.predict(user_id, item_id)
    return jsonify({'user_id': user_id, 'item_id': item_id, 'predicted_rating': prediction.est})

@app.route('/forecast', methods=['GET'])
def forecast():
    forecast = forecast_model.predict(future)
    forecast_json = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_dict('records')
    return jsonify(forecast_json)

if __name__ == '__main__':
    app.run(debug=True)
```
